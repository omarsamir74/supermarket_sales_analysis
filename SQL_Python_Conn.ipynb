```python
#What is the total sales amount for each branch and city?
analysis_1 = '''SELECT b.Branch_name, b.City, SUM(s.Total) AS Total_sales FROM Branch_Dimension AS b
INNER JOIN Sales_Fact as s
ON b.Branch_ID = s.Branch_ID
GROUP BY  b.Branch_name, b.City
ORDER BY Total_sales;
'''
```

```python
cursor = conn.cursor()
cursor.execute(analysis_1)
for row in cursor:
    print(row)
```

```python
#Which product line has the highest average customer rating?
analysis_2 = '''SELECT Product_Line, AVG(rating) AS Rate FROM Product_Dimension AS P
INNER JOIN Sales_Fact AS S
ON P.product_ID = S.product_ID
GROUP BY Product_Line
ORDER BY Rate DESC;
'''
```

```python
cursor = conn.cursor()
cursor.execute(analysis_2)
for row in cursor:
    print(row)
```

```python
#How do total sales vary by payment method?
analysis_3 = '''SELECT Payment, SUM(Total) AS Total_sales FROM Payment_Dimension AS P
JOIN Sales_Fact AS S
ON P.Payment_ID = S.Payment_ID
GROUP BY Payment
ORDER BY Total_sales ASC;
'''
```

```python
cursor = conn.cursor()
cursor.execute(analysis_3)
for row in cursor:
    print(row)
```

```python
#What is the total sales amount for each customer type?
analysis_4 = '''SELECT  Customer_Type, SUM(Total) AS Total_sales FROM Customer_Dimension AS c
JOIN Sales_Fact AS S
ON c.Customer_ID = S.Customer_ID
GROUP BY Customer_Type
ORDER BY Total_sales desc;
'''
```

```python
cursor = conn.cursor()
cursor.execute(analysis_4)
for row in cursor:
    print(row)
```

```python
#How do sales numbers vary across different years ?
analysis_5 = '''SELECT  Year, SUM(Total) AS Total_sales FROM Time_Dimension AS T
JOIN Sales_Fact AS S
ON T.ID = S.ID
GROUP BY Year
ORDER BY Total_sales ;
'''
```

```python
cursor = conn.cursor()
cursor.execute(analysis_5)
for row in cursor:
    print(row)
```

```python
#Which are the top 2 branches with the highest total sales?
analysis_6 = '''SELECT TOP(2) Branch_name, SUM(Total) AS Total_sales FROM Branch_Dimension AS B
JOIN Sales_Fact AS S
ON B.Branch_ID = S.Branch_ID
GROUP BY Branch_name
ORDER BY Total_sales ;
'''
```

```python
cursor = conn.cursor()
cursor.execute(analysis_6)
for row in cursor:
    print(row)
```

```python
#What is the total sales amount for each customer loyalty status?
analysis_7 = '''
SELECT Loyalty_Status, SUM(Total) AS Total_sales FROM Customer_Dimension AS C
JOIN Sales_Fact AS S
ON C.Customer_ID = S.Customer_ID
GROUP BY Loyalty_Status
ORDER BY Total_sales ;
'''
```
